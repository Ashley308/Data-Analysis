##### To calculate the LOF you need to apply the following steps:
 1. Calculate the distance between each pair of points p1,p2
 2. Find your Kth nearest neighbor point (x) from current point (p):
    - After finding it, remember the distance *Nd* which is the distance from p to x, and save a set of elements which are in a neighborhood of *Nd* around point p.
    - Save the length of this set of elements.
 3. Calculate ReachDistance(x,p)=max(Nd_x, Distance(x,p))
 3. Calculate LRD (Local Reachability Distance) where LRD_p = length_of_set_p/Sum[ReachDistance(x,p)] (for each x in set of elements of point p)
 4. Calculate LOF (Local Outlier Factor) where LOF_p = {Sum[LRD_x] (for each x in set of elements of point p) * Sum[RD(x,p)] (for each x in set of elements of point p)} / length_of_set_p^2

```python
## Example input
list_of_points = [[0,0],[-1,0],[0,-1],[1,0],[5,5]]
# get_outliers(list_of_points,2,distance_approach="Manhattan") ==> [(5,5)]
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import iqr

def calculate_point_to_point_distance(point_a, point_b, distance_approach="Euclidean"):
    if distance_approach == "Euclidean":
        return np.sqrt(np.sum((np.array(point_a) - np.array(point_b))**2))
    elif distance_approach == "Manhattan":
        return np.sum(np.abs(np.array(point_a) - np.array(point_b)))
    else:
        raise ValueError("Unsupported distance approach")

def get_k_nearest_neighbors(point, list_of_points, K, distance_approach):
    distances = [calculate_point_to_point_distance(point, p, distance_approach) for p in list_of_points if p != point]
    sorted_distances = sorted(distances)
    k_distance = sorted_distances[K-1]
    neighbors = [p for p in list_of_points if calculate_point_to_point_distance(point, p, distance_approach) <= k_distance and p != point]
    return neighbors, k_distance

def local_reachability_density(point, neighbors, k_distances, distance_approach):
    reach_distances = [max(k_distances[tuple(neighbor)], calculate_point_to_point_distance(point, neighbor, distance_approach)) for neighbor in neighbors]
    return len(neighbors) / sum(reach_distances)

def get_points_lof(list_of_points, K=2, distance_approach="Euclidean"):
    k_distances = {}
    neighbors = {}
    for point in list_of_points:
        neighbors[tuple(point)], k_distances[tuple(point)] = get_k_nearest_neighbors(point, list_of_points, K, distance_approach)
    
    lrds = {tuple(point): local_reachability_density(point, neighbors[tuple(point)], k_distances, distance_approach) for point in list_of_points}
    
    lof_scores = {}
    for point in list_of_points:
        lof = sum(lrds[tuple(neighbor)] / lrds[tuple(point)] for neighbor in neighbors[tuple(point)]) / len(neighbors[tuple(point)])
        lof_scores[tuple(point)] = lof
    
    return lof_scores

def get_outliers(list_of_points, K=2, distance_approach="Euclidean"):
    lof_scores = get_points_lof(list_of_points, K, distance_approach)
    
    # Calculate box plot statistics
    lof_values = list(lof_scores.values())
    q1 = np.percentile(lof_values, 25)
    q3 = np.percentile(lof_values, 75)
    iqr_value = iqr(lof_values)
    upper_bound = q3 + 1.5 * iqr_value
    
    # Identify outliers
    outliers = [point for point, score in lof_scores.items() if score > upper_bound]
    
    # Plot box plot
    plt.figure(figsize=(10, 6))
    plt.boxplot(lof_values)
    plt.title("Box Plot of LOF Scores")
    plt.ylabel("LOF Score")
    plt.show()
    
    return outliers

# Example usage
list_of_points = [[0,0],[-1,0],[0,-1],[1,0],[5,5]]
outliers = get_outliers(list_of_points, K=2, distance_approach="Manhattan")
print("Outliers:", outliers)
```

The major routines comprising this implementation include the following: calculate_point_to_point_distance: returns the distance between two points in either Euclidean or Manhattan distance. get_k_nearest_neighbors: returns the K nearest neighbors around a given point. 

local_reachability_density: calculates the local reachability density for a given point. get_points_lof: implements LOF, returning a dictionary of LOF scores for each point.

get_outliers: This uses LOF scores in order to decide on outliers by using a box plot method; it also visualizes the box plot of LOF scores.
In the get_outliers function, the IQR method is used to decide on the threshold. Points whose LOF scores are higher than Q3 + 1.5 * IQR will be decided as outliers.
